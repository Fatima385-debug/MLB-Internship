```python
SUBJECTS = ["Python", "Mathematics", "Statistics", "Machine_Learning"]
```

```python
subject_averages = {subject: np.mean(df[subject]) for subject in SUBJECTS}
for subject, avg in subject_averages.items():
    print(f"{subject:20s}: {avg:.2f}")
```

```python
df["Total_Marks"] = df[SUBJECTS].sum(axis=1)
df["Average_Marks"] = np.round(df[SUBJECTS].mean(axis=1), 2)
```

```python
overall_avg = np.mean(df["Average_Marks"])
print(f"Overall class average: {overall_avg:.2f}")
df.head()
```

```python
top_5 = df.sort_values(by="Average_Marks", ascending=False).head(5)
top_5[["Student_ID", "Name", "Average_Marks", "Total_Marks"]]
```

```python
below_average = df[df["Average_Marks"] < overall_avg]
below_average[["Student_ID", "Name", "Average_Marks"]]
```

```python
print(f"Number of students below average: {len(below_average)}")
```

```python
total_students = df["Student_ID"].nunique()
print(f"Total number of students: {total_students}")
```

```python
df_sorted = df.sort_values(by="Average_Marks", ascending=False)
df_sorted.to_csv("processed_analysis.csv", index=False)
print("Saved: processed analysis.csv")
```
