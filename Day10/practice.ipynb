Practice 3 — Experiment with Different Activation Functions

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
from tensorflow import keras
from tensorflow.keras import layers

def build_model(activation_name):
    model = keras.Sequential([
        layers.Input(shape=(4,)),
        layers.Dense(8, activation=activation_name, name='hidden_layer'),
        layers.Dense(1, activation='sigmoid', name='output_layer')
    ])
    return model

for act in ['relu', 'sigmoid', 'tanh']:
    print(f"\n--- Model with '{act}' activation in hidden layer ---")
    model = build_model(act)
    model.summary()
    # Explicitly print the activation function used in the hidden layer
    hidden_activation = model.get_layer('hidden_layer').activation.__name__
    print(f"Hidden layer activation function: {hidden_activation}")
```
